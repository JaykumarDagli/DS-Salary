### Checking ratings of the jobs 

```python
df['Rating'].hist(grid=False)
```

Histogram is right skewed with many ratings with value greater than 3.5

### Avg Salary of the Jobs 

```python
df.avg_salary.hist(bins=6,grid=True)
```

There are many jobs in the range of 60K - 80K, the number reduces after 80K and again increases between 120K - 140K.

### Age of company vs Jobs offered

```python
df.age.hist(bins=15,grid=False)
```

Most of the jobs are offered by companies who were recently founded within the last ~30 years

### What is the length of the Job description 

```python
df.len_desc.hist(grid=False)
```

Majority of the job description contains 3500-4500 words; but some also have a detailed explanation of 12000-14000 words

### Comparing the age of the company and job desciption with salary 

```python
sns.scatterplot(x=df.len_desc,y=df.age,hue=df.avg_salary)
```

The really long descriptions are jobs that probably won't pay as much as the others.
The really old companies (200+ years) actually keep their descriptions short, and also pay on a higher scale typically.

### Checking the outliers using boxplots 

```python
df.boxplot(column=['age','avg_salary'])
```

```python
df.boxplot(column='Rating')
```

The boxplots show that age has a lot of outliers on the higher side, rating has a few on the lower side, while average salary has a very uniform distribution.

### Checking co-relation with heatmap

```python
c1=df[['age','avg_salary','Rating','len_desc','num_comp']].corr()
c1
```

```python
sns.heatmap(c1)
```

There is slight co-relation between average salary with age of company and Ratings, rest are not co-related

### Pairplots 

```python
sns.pairplot(df[['age','avg_salary','Rating','len_desc','num_comp']])
```

### Categorical columns 

```python
df_cat=df[['Location','Headquarters','Size','Type of ownership','Industry','Sector','Revenue','Company Name',
           'State','python_yn','r_yn','aws_yn','spark_yn', 'hadoop_yn', 'docker_yn',
       'sql_yn', 'linux_yn', 'flask_yn', 'django_yn', 'tensorflow_yn',
       'keras_yn', 'pytorch_yn', 'algo_yn', 'stats_yn', 'tableau_yn',
       'job_type', 'seniority']]
```

```python
for i in df_cat.columns:
    cat_num = df_cat[i].value_counts()
    print("graph for %s: total = %d" % (i, len(cat_num)))
    chart = sns.barplot(x=cat_num.index, y=cat_num)
    chart.set_xticklabels(chart.get_xticklabels(), rotation=90)
    plt.show()
```

### Choosing Top 20 for location, hq, industry and company 

```python
for i in df[['Location','Headquarters','Industry','Company Name']].columns:
    cat_num = df_cat[i].value_counts()[:20]
    print("graph for %s: The Top %d" % (i, len(cat_num)))
    chart = sns.barplot(x=cat_num.index, y=cat_num)
    chart.set_xticklabels(chart.get_xticklabels(), rotation=90)
    plt.show()
```

### Checking Salary according to Job Type

```python
pd.pivot_table(df,index='job_type',values='avg_salary',aggfunc=[np.min,np.max,np.mean,np.std]).sort_values('job_type',ascending=False)
```

Data scientist have avg salary of 116K, data analyst and engineers have abg salary of ~100K

### Checking salary as per seniority type

```python
pd.pivot_table(df,index=['job_type','seniority'],values='avg_salary')
```

### Checking salary state wise 

```python
pd.pivot_table(df,index='State',values='avg_salary',margins=True,aggfunc=[np.mean,'count'])
```

New Hamshire offers mean avg salary of 148K, highest in the USA
